```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

plt.rcParams['figure.figsize'] = (18, 10)
```

```python
var = np.linspace(0,3.3,55000)
dig = np.round(var/3.3*1024)
```

```python
plt.tick_params(labelsize=20)
plt.plot(var[22500:23000], dig[22500:23000], '-', linewidth=2.0)
plt.xlabel('Real Voltage (V)', fontsize=24)
plt.ylabel('Measured value (bits)', fontsize=24)
```
